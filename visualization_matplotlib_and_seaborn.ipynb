# figure size DPI

```python
#here dpi is dot per inch or pixsels per inch and figsize=(width ,height)
fig=plt.figure(figsize=(3,2))
ax=fig.add_axes([0,0,1,1])
x=np.arange(1,11)
y=x**2
ax.plot(x,x**2,'b--',label='x squre')
ax.plot(x,x**3,'y--',label ='x  qube')

# how to add lagneds in graph 
ax.legend() # and we also need to add lagends in plot ax as above
```

```python
fig.savefig('my file.png',dpi=100)# here  it is used to save figure 
```

```python
#here dpi is dot per inch or pixsels per inch and figsize=(width ,height)
fig=plt.figure(figsize=(3,2))
ax=fig.add_axes([0,0,1,1])
x=np.arange(1,11)
y=x**2
ax.plot(x,x**2,'b--',label='x squre')
ax.plot(x,x**3,'y--',label ='x  qube')

# how to add lagneds in graph 
ax.legend(loc=0) # and we also need to add lagends in plot ax as above
```

```python
fig=plt.figure(figsize=(5,4))
new_ax=fig.add_axes([0,0,.9,.9])
new_ax.set_xlabel('X axis')
new_ax.set_ylabel('Y axis')
new_ax.set_title('Title ')
new_ax.plot(x,y,color='orange') #here we can use rgb
new_ax.plot(y,x,color='purple',linewidth=4,linestyle='steps')
new_ax.plot(x,x*x ,'r*-')
```

# Seaborn library practice

```python
import numpy as np 
import pandas as pd 
from pandas import DataFrame ,Series
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
tips=sns.load_dataset('tips')
tips.head()
```

```python
sns.distplot(tips['total_bill'])
plt.grid()
```

```python
sns.distplot(tips['total_bill'],kde=False ,bins=30)
plt.grid()
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips ,kind='hex')
plt.grid()
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips ,kind='reg')
plt.grid()
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips ,kind='kde')
plt.grid()
```

```python
sns.pairplot(tips ,hue='sex',palette='coolwarm') #it will make all possible pair of numaric value with given value
plt.grid()
```

```python
sns.rugplot(tips['total_bill'])
```

```python
sns.kdeplot(tips['total_bill'])
```
