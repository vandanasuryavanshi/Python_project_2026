```python
costumer_review=["sam was great help to me in the store",
                "the cashier was very rude to me ,I think her was eleanor",
                "amazing work from sadeen!",
                "sarah helped me to find the items I needed quickly",
                "lucky is such a great addition to the team",
                "great service from sara she found me what i wanted"]
costumer_review
```

```python
## create code to find all reviews from sarah
Sarahs_review=[]
for review in costumer_review:
    if re.search(r"sarah?",review):
        Sarahs_review.append(review)
        
        
Sarahs_review
```

```python
#create review all starting from a
a_string=[]
for review in costumer_review:
    if re.search(r"^a",review):
        a_string.append(review)
        
a_string
```

```python
#find all review ending with letter y
y_end=[]
for review in costumer_review:
    if re.search(r"y$",review):
        y_end.append(review)
    
y_end
```

```python

need_want_review=[]
for review in costumer_review:
    if (re.search(r"(need|want)ed",review)):
        need_want_review.append(review)
        
need_want_review
        
```

```python
## no punctuantion 
no_punc=[]
for review in costumer_review:
    no_punc_string=re.sub(r'[^\w\s]',"",review)
    no_punc.append(no_punc_string)
no_punc
```

# Stemming :- reducing words to base root form ex connected will connect
here we remove suffix so some time get new unique words in steming we try to have reduce uinique words

```python
import nltk
from nltk.stem import PorterStemmer #it reduces english work make it easier from bigger complex



```

```python
ps=PorterStemmer()
ps
```

```python
connect_tokens=["connected","connecting","connectivity",'connects',"connect","connector"]
connect_tokens
```

```python
for word in connect_tokens:
    print(word, " : " , ps.stem(word))
```

```python
learn_token=["learning","learn","learned","learner"]
for word in learn_token:
    print(f" {word} : {ps.stem(word)}")
```

# lemmantization reduce the word to meaningfull small word and store while preserving meaning
lemmatitzation used prediend dictionary words to keep the context

```python
nltk.download("wordnet")
from nltk.stem import WordNetLemmatizer
```

```python
lemmatizer=WordNetLemmatizer()
```

```python
for word in connect_tokens:
    print(word ,": ",lemmatizer.lemmatize(word))
```

```python
words=["like",'test','tester',"rocky","shinny",'biiter']
for word in words:
    print(word,":",lemmatizer.lemmatize(word))
```
